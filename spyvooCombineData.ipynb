```python
import pandas as pd
import datetime as dt
```

```python
spy = pd.read_csv('./data/spy.csv')
voo = pd.read_csv('./data/voo.csv')
```

```python
spy.tail()
```

```python
voo.tail()
```

```python
spy = spy[['DATE', 'TIME_M', 'PRICE']]
voo = voo[['DATE', 'TIME_M', 'PRICE']]
```

```python
spy.tail()
```

```python
def toDateTime(date, time):
    year = date // 10000
    month = date // 100 % 100
    day = date % 100
    time = time.split(":")
    h = int(time[0])
    m = int(time[1])
    s = float(time[2])
    ms = (round(float(time[2]), 6) - int(s)) * 1000000
    if int(ms) == 1000000:
        ms = 0
        if int(s) == 59:
            s = 0
            if m == 59:
                m = 0
                h += 1
            else:
                m += 1
        else:
            s += 1
    return dt.datetime(year, month, day, hour=h, minute=m, second=int(s), microsecond=int(ms))
```

```python
spy['datetime'] = [toDateTime(spy['DATE'].iloc[i], spy['TIME_M'].iloc[i]) for i in range(len(spy))]
spy['datetime'].tail()
```

```python
voo['datetime'] = [toDateTime(voo['DATE'].iloc[i], voo['TIME_M'].iloc[i]) for i in range(len(voo))]
voo['datetime'].tail()
```

```python
spy = spy.set_index('datetime')
spy = spy.rename(columns={'PRICE': 'spyprice'})
spy = spy[['spyprice']]
spy.tail()
```

```python
voo = voo.set_index('datetime')
voo = voo.rename(columns={'PRICE': 'vooprice'})
voo = voo[['vooprice']]
voo.tail()
```

```python
data = spy.join(voo, how='outer')
data = data.ffill()
data.tail(50)
```
